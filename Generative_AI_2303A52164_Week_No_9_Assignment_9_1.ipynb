# **Dataset: MNIST**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adadelta
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

```python
# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Normalize the data
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
# Convert labels to categorical
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# Define ANN model
def build_model(hidden_layers=[32, 32, 32], activation='relu', optimizer=Adadelta(), epochs=10, batch_size=128):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
 # Add hidden layers
    for neurons in hidden_layers:  # Indentation corrected here
        model.add(Dense(neurons, activation=activation))
    return model # Added return statement to complete the function
```

```python
model.add(Dense(10, activation='softmax'))
```

```python
def evaluate_model(history):
    """Evaluate the model and return the model, training accuracy, and testing accuracy."""
    train_acc = history.history['accuracy'][-1]
    test_acc = history.history['val_accuracy'][-1]
    return model, train_acc, test_acc  # Assuming 'model' is defined globally
```

```python
# Tune architecture for best accuracy
tuned_architectures = [
    ([64, 64, 64], 'relu'),
    ([128, 128, 128], 'relu'),
    ([256, 128, 64], 'relu'),
    ([64, 64], 'tanh'),
    ([128, 64], 'tanh'),
]
```

```python
best_acc = 0
best_model = None
best_params = None
```

```python
for layers, activation in tuned_architectures:
    print(f"Testing architecture: Layers={layers}, Activation={activation}")
    model = build_model(hidden_layers=layers, activation=activation) # Only get the model
for layers, activation in tuned_architectures:
    print(f"Testing architecture: Layers={layers}, Activation={activation}")
    model = build_model(hidden_layers=layers, activation=activation) # Only get the model
    model.add(Dense(10, activation='softmax')) # Add the output layer
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # Compile the model
    history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10)
```

```python
if test_acc > best_acc:
        best_acc = test_acc
        best_model = model
        best_params = (layers, activation)
```

```python
print(f"Best Architecture: Layers={best_params[0]}, Activation={best_params[1]}")
print(f"Best Testing Accuracy: {best_acc:.4f}")
```
